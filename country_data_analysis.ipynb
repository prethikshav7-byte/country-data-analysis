Task 1 — Fetch and Validate the Response

```python
import requests

url = url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"

try:
    response = requests.get(url)

    print("Status Code:", response.status_code)

    if response.status_code == 200:
        data = response.json()
        print("Total countries:", len(data))
    else:
        print("Error:", response.text)

except Exception as e:
    print("Error occurred:", e)
```

Task 2 — Extract and Display Key Fields

```python
import pandas as pd

countries_list = []

for country in data:
    name = country.get('name', {}).get('common', 'N/A')
    population = country.get('population', 0)
    region = country.get('region', 'Unknown')
    area = country.get('area', 0)

    countries_list.append({
        "common_name": name,
        "population": population,
        "region": region,
        "area": area
    })

# Create DataFrame
df = pd.DataFrame(countries_list)

# Display first 10 rows
df.head(10)
```

Task 3 — Summarize the Data

Region with Highest Total Population

```python
# Make sure numeric columns are clean
df["population"] = pd.to_numeric(df["population"], errors="coerce")
df["area"] = pd.to_numeric(df["area"], errors="coerce")

# Fill missing values
df["population"] = df["population"].fillna(0)
df["area"] = df["area"].fillna(0)

# Group by region
region_population = df.groupby("region")["population"].sum()

# Get max
highest_region = region_population.idxmax()
highest_value = region_population.max()

print("Region with Highest Total Population:")
print(highest_region, "-", highest_value)
```

Country with Largest Area

```python
# Drop rows where area is 0 (optional but safer)
df_clean = df[df["area"] > 0]

largest_country = df_clean.loc[df_clean["area"].idxmax()]

print("\nCountry with Largest Area:")
print(largest_country["common_name"], "-", largest_country["area"])
```
